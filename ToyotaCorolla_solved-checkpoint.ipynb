## preparing a model 

```python
import statsmodels.formula.api as smf 
model = smf.ols('Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight', data=ToyotaCorolla2).fit()
```

## Calculating VIF

```python
# 1) Collinearity Problem Check
# Calculate VIF = 1/(1-Rsquare) for all independent variables

rsq_Age_08_04 = smf.ols('Age_08_04~KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight', data = ToyotaCorolla2).fit().rsquared
vif_Age_08_04 = 1/(1-rsq_Age_08_04)


rsq_KM = smf.ols('KM~Age_08_04+HP+cc+Doors+Gears+Quarterly_Tax+Weight', data = ToyotaCorolla2).fit().rsquared
vif_KM = 1/(1-rsq_KM)


rsq_HP = smf.ols('HP~Age_08_04+KM+cc+Doors+Gears+Quarterly_Tax+Weight', data = ToyotaCorolla2).fit().rsquared
vif_HP = 1/(1-rsq_HP)


rsq_cc = smf.ols('cc~Age_08_04+HP+KM++Doors+Gears+Quarterly_Tax+Weight', data = ToyotaCorolla2).fit().rsquared
vif_cc = 1/(1-rsq_cc)


rsq_Doors = smf.ols('Doors~Age_08_04+HP+cc+KM+Gears+Quarterly_Tax+Weight', data = ToyotaCorolla2).fit().rsquared
vif_Doors = 1/(1-rsq_Doors)


rsq_Gears = smf.ols('Gears~Age_08_04+HP+cc+Doors+KM+Quarterly_Tax+Weight', data = ToyotaCorolla2).fit().rsquared
vif_Gears = 1/(1-rsq_Gears)


rsq_Quarterly_Tax = smf.ols('Quarterly_Tax~Age_08_04+HP+cc+Doors+Gears+KM+Weight', data = ToyotaCorolla2).fit().rsquared
vif_Quarterly_Tax = 1/(1-rsq_Quarterly_Tax)
rsq_Weight = smf.ols('Weight~Age_08_04+HP+cc+Doors+Gears+KM+Quarterly_Tax', data = ToyotaCorolla2).fit().rsquared
vif_Weight = 1/(1-rsq_Weight)


#storing the values in a dataframe
D1 = {'variables':['Weight','Age_08_04', 'HP' ,'cc' ,'Doors' ,'Gears' ,'KM', 'Quarterly_Tax' ], 'vif':[vif_Weight,vif_Age_08_04,vif_HP,vif_cc,vif_Doors,vif_Gears,vif_KM,vif_Quarterly_Tax]}
vif_frame= pd.DataFrame(D1)
vif_frame
```

```python
# None variable has VIF>20, No Collinearity, so consider all varaibles in Regression equation
```

## Residual Plot for Homoscedasticity

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

## Detecting Influencers/Outliers

## Cook’s Distance

```python
# Get influencers using cook's distance
model_influence = model.get_influence()
(c,_) = model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize = (20,7))
plt.stem(np.arange(len(ToyotaCorolla2)), np.round(c,3))
plt.xlabel('Row Index')
plt.ylabel('cooks distance')
plt.show()
```

```python
#index and value of influencer where c is more than .5
(np.argmax(c),np.max(c))
```
