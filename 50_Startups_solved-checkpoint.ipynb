```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
startups = pd.read_csv('50_Startups.csv')
startups
```

```python
startups.info()
```

```python
# creating a copy of this original dataset
startups1 = startups.copy()
```

## datatype  conversion

```python
startups1['State']=startups1['State'].astype('category')
```

```python
startups1.info()
```

## duplicates

```python
startups1[startups1.duplicated()].shape
```

```python
#renaming
startups2 = startups1.rename({'R&D Spend':'RD_Spend', 'Marketing Spend':'Marketing_Spend'},axis=1)
```

```python
startups2
```

## outliers detection


```python
#histogram
startups2['Profit'].hist()
```

```python
startups2.boxplot(column=['Profit'])
```

```python
startups2['Profit'].describe()
```

```python
#correlation matrix
startups2.corr()
```

## scatterplot between variables

```python
sns.set_style(style='darkgrid')
sns.pairplot(startups2)
```

## preparing a model

```python
import statsmodels.formula.api as smf
model = smf.ols('Profit~RD_Spend+Administration+Marketing_Spend', data=startups2).fit()
```

```python
model.summary()
```

```python
model.params
```

```python
#tvalue and pvalue
print(model.tvalues, '\n', model.pvalues)
```

```python
#r-squared and adj_r-squared
(model.rsquared,model.rsquared_adj)
```

## Model predictions

```python

new_data=pd.DataFrame({'RD_Spend':70000,"Administration":90000,"Marketing_Spend":140000},index=[0])
new_data
```

```python
model.predict(new_data)
```

```python
pred_y=model.predict(startups2)
pred_y
```
